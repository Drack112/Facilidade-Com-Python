# Visualização de dados

```python
import pandas as pd
```

```python
numero_dias = 1000
datas = pd.date_range(start='1/1/2023', periods=numero_dias)
```

```python
df = pd.DataFrame(range(numero_dias), columns=['number'], index=datas)
df.index.name = 'datas'
```

```python
df
```

```python
df.plot()
```

```python
df2 = df.copy()
df2['number2'] = df['number'] *30
```

```python
df2['number'].plot()
```

```python
df2.plot()
```

```python
df2.iloc[:5].plot(kind='bar')
```
