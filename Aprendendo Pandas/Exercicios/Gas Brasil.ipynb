## Converter as datas para Timestamp

```python
df['DATA INICIAL'] = pd.to_datetime(df['DATA INICIAL'])
df['DATA FINAL'] = pd.to_datetime(df['DATA FINAL'])
```

## Representar uma novo coluna com ano e o mês utilizando a coluna DATA FINAL como referência

```python
df['ANO-MES'] = df['DATA FINAL'].apply(lambda x: "{}".format(x.year)) + df["DATA FINAL"].apply(lambda x: "/{:02d}".format(x.month))
```

## Qual o preço médio da revenda de gasolina em agosto de 2008?

```python
df2[df2['ANO-MES'] == '2008/08']['PREÇO MÉDIO REVENDA'].mean()
```

## Obter uma tabela contendo a variação percentual ano a ano para o estado do RJ

```python
df2["MES"] = df2['DATA FINAL'].apply(lambda x: x.month)
df_rio = df2[df2["ESTADO"] == "RIO DE JANEIRO"]
```

```python
df_month_rio = df_rio.groupby("ANO-MES")[["PREÇO MÉDIO REVENDA", "MES"]].last()
```

```python
df_month_rio[df_month_rio["MES"] == 12]
```

```python
((df_month_rio[df_month_rio["MES"] == 12] / df_month_rio[df_month_rio["MES"] == 12].shift(1) - 1) *100).rolling(3).sum()
```

## Criar uma  tabela contendo uma série temporal com a diferença absoluta e percentual entre os valores mais baratos e caros

```python
df_max = df2.groupby("ANO-MES").max()["PREÇO MÉDIO REVENDA"]
df_min = df2.groupby("ANO-MES").min()["PREÇO MÉDIO REVENDA"]
```

```python
df_diff = pd.DataFrame()
```

```python
df_diff
```

```python
df_diff["abs_diff"] = df_max - df_min
df_diff["percent_diff"] = (df_max - df_min) / df_min + 100
```
